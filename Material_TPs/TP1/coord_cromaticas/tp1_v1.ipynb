```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# Cargar una imagen
img2 = cv.imread('CoordCrom_2.png', )
```

```python
# Vemos la imagen original
plt.imshow(img2);
```

```python
def cambio_coord_crom(img):
    # Se realiza un cambio de coordenadas a RGB
    img = img[...,::-1]
    
    # Realizar un split en canales
    img_r, img_g, img_b = cv.split(img)
    
    # Castear a float32
    img_r = img_r.astype('float32')
    img_g = img_g.astype('float32')
    img_b = img_b.astype('float32')
    
    # Sumar los 3 canales para usarlos como divisor
    color_sum = img_r + img_g + img_b

    # Crear imagen modificada con el cambio de coordenadas    
    img_r_mod = np.divide(img_r, color_sum)
    img_g_mod = np.divide(img_g, color_sum)
    img_b_mod = np.divide(img_b, color_sum)

    # Creamos una merge de las coordenadas
    img_mod = cv.merge((img_r_mod.astype('uint8'), img_g_mod.astype('uint8'), img_b_mod.astype('uint8')))

    # Clipeamos para evitar la saturación
    np.clip(img_mod, 0, 255, out=img_mod)
    
    return img_mod
```

```python
# Implementamos esta función para las otras dos figuras
img1_mod = cambio_coord_crom(cv.imread('CoordCrom_1.png'))
plt.figure()
plt.imshow(img1_mod)
plt.title('Figura 1 modificada')
plt.show()
```

```python
# Implementamos esta función para las otras dos figuras
img1_mod = cambio_coord_crom(cv.imread('CoordCrom_2.png'))
plt.figure()
plt.imshow(img1_mod)
plt.title('Figura 2 modificada')
plt.show()
```

```python
img3_mod = cambio_coord_crom(cv.imread('CoordCrom_3.png'))
plt.figure()
plt.imshow(img3_mod)
plt.title('Figura 3 modificada')
plt.show()
```
